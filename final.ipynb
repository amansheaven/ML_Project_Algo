```python
from sklearn import linear_model
import tkinter as tk 
import statsmodels.api as sm
import pandas as pd
import numpy as np
import requests
import json
import io
url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=AAPL&outputsize=full&apikey=47DJ9RC2J7RJYJ9B&datatype=csv"

s=requests.get(url).content

df=pd.read_csv(io.StringIO(s.decode('utf-8')))
print(df)
 

X = df[['open','high','low','volume']] 
Y = df['close'] 
regr = linear_model.LinearRegression()
regr.fit(X, Y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)



X = sm.add_constant(X) # adding a constant
 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
root= tk.Tk() 
 
canvas1 = tk.Canvas(root, width = 1400, height = 800,bg='lightblue')
canvas1.pack()


Intercept_result = ('Intercept: ', regr.intercept_)
label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center',bg='lightblue')
canvas1.create_window(260, 220, window=label_Intercept)

Coefficients_result  = ('Coefficients: ', regr.coef_)
label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center',bg='lightblue')
canvas1.create_window(260, 240, window=label_Coefficients)


print_model = model.summary()
label_model = tk.Label(root, text=print_model, justify = 'center', relief = 'solid', bg='Lightgreen')
canvas1.create_window(900, 240, window=label_model)

label1 = tk.Label(root, text='OPEN VALUE ',bg='lightblue')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) 
canvas1.create_window(270, 100, window=entry1)

label2 = tk.Label(root, text='LOW VALUE',bg='lightblue')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root) 
canvas1.create_window(270, 120, window=entry2)

label3 = tk.Label(root, text='HIGH VALUE ',bg='lightblue')
canvas1.create_window(100, 140, window=label3)


entry3 = tk.Entry (root) 
canvas1.create_window(270, 140, window=entry3)

label4 = tk.Label(root, text='VOLUME ',bg='lightblue')
canvas1.create_window(100, 160, window=label4)

entry4 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 160, window=entry4)


def values(): 
    global open_value 
    open_value = float(entry1.get()) 
    
    global low_value 
    low_value = float(entry2.get()) 
    
    
    global high_value
    high_value=float(entry3.get())
    
    global volume_value
    volume_value=float(entry4.get())
    
    Prediction_result  = ('Predicted Closing Value: ', regr.predict([[open_value ,low_value,high_value,volume_value]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='yellow')
    canvas1.create_window(270, 380, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Stock Index Price',command=values, bg='grey') # button to call the 'values' command above 
canvas1.create_window(270, 300, window=button1)
 

root.mainloop()
```
